Step 7: Here is the rest of the code where 

1. We are using an infinite while loop so that we can chat with the ChatGPT API repeatedly without executing the code again and again. 
2. In the second line we a taking input from the user and store it in a variable ‘message’.
3. If a user inputs any question then only we enter the if condition and make a JSON of file and append it to the JSON file that we have created in step 6 after that generate the chat using openai.ChatCompletion.create()
4. Store the answer in the variable ‘reply’ and print that reply using the print() function.

```python
import openai
openai.api_key = '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'
messages = [ {"role": "system", "content":
			"You are a intelligent assistant."} ]
while True:
	message = input("User : ")
	if message:
		messages.append(
			{"role": "user", "content": message},
		)
		chat = openai.ChatCompletion.create(
			model="gpt-3.5-turbo", messages=messages
		)
	reply = chat.choices[0].message.content
	print(f"ChatGPT: {reply}")
	messages.append({"role": "assistant", "content": reply})
```
